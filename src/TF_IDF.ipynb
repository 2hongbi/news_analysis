```python
'''
kkma = Kkma()
okt = Okt()
komoran = Komoran()
twitter = Twitter()
'''
hannanum = Hannanum()
```

```python
len(comment)
```

```python
comment_tag=[]
n=0;
for i in range(0,1510):
    a=hannanum.nouns(comment[i])
    a_join=" ".join(a)
    comment_tag.append(a_join)
    print(n,":",a_join)
    n+=1
```

```python
print(comment_tag)
```

# TF-IDF : scikit-learn

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector = CountVectorizer()
print(vector.fit_transform(comment_tag).toarray()) 
print(vector.vocabulary_)
```

```python
tfidfv = TfidfVectorizer().fit(comment_tag)
print(tfidfv.transform(comment_tag).toarray())
print(tfidfv.vocabulary_)
```

```python
tfidfv.vocabulary_
```

```python
for (key, value) in sorted(tfidfv.vocabulary_.items(), reverse=True):
    print(key, ":", value)
```
